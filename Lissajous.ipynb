<h1>Lissajous Patterns<h1>
$$x = A \sin (at+\delta)$$
$$y = B \sin (bt)$$

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def lissajous(cycles, step_size = 0.01, a, delta, b):
    """
    one cycle is 2 pi radians
    step size is the fineness of calculation, default is 0.01
    The amplitudes of both x and y sine waves are fixed at one
    a is the frequency of the sine wave in x axis
    b is the frequency of the sine wave in y axis
    delta is the phase difference between the two sine waves
    
    """
    t = np.arange(0, 2*cycles*np.pi, step_size)
    
```

```python
t = np.arange(0, 10*np.pi, 0.01)
# parameters for x
A = 1*np.pi # amplitude of x
a = 1 # frequency  of x
delta = 2*np.pi/4 # phase difference np.pi/2 for circle
# parameters for y
B = 1*np.pi # amplitude of y
b = 2 # frequency of y
# calculate x and y
x = A*np.sin(a*t + delta)
y = B*np.sin(b*t)
```

```python
%matplotlib inline
#change inline to notebook for interactive plot
unit = np.pi/4
x_tick_labels = [r"$-\pi$", "", r"$-\frac{\pi}{2}$", "", r"$0$", "", r"$\frac{\pi}{2}$", "", r"$\pi$"]
plt.figure(figsize=(7, 7))
plt.axes(aspect='equal')
plt.plot(x, y)
plt.xticks(np.arange(-np.pi, np.pi+unit, step=unit), x_tick_labels, fontsize=14, rotation=0)
plt.xlabel(r"$\omega$")
plt.grid(b=True, which='major', axis='both')
```

```python
%matplotlib notebook
fig, ax = plt.subplots(figsize=(7, 7))
ax.set(xlim=(-np.pi, np.pi), ylim=(-np.pi, np.pi), aspect='equal')

_ = plt.plot(x, y)

redDot, = plt.plot(x[0], y[0], 'ro')
blueDot, = plt.plot(x[0], y[0], 'bo')
greenDot, = plt.plot(x[0], y[0], 'go')

def animate1(i):
    redDot.set_data(A*np.sin(a*i + delta), B*np.sin(b*i))
    blueDot.set_data(-np.pi, B*np.sin(b*i))
    greenDot.set_data(A*np.sin(a*i + delta), -np.pi)
    return redDot, blueDot, greenDot,

def animate2(i):
    redDot.set_data(-A*np.sin(a*i + delta), -B*np.sin(b*i))
    blueDot.set_data(-np.pi, -B*np.sin(b*i))
    greenDot.set_data(-A*np.sin(a*i + delta), -np.pi)
    return redDot, blueDot, greenDot,

# create animation using the animate() function
myAnimation = animation.FuncAnimation(fig, animate1, frames=t, \
                                      interval=5, blit=False, repeat=False)

plt.show()

# To save the animation
#writer = animation.FFMpegWriter(fps=15, metadata=dict(artist='Me'), bitrate=1800)
#_ = myAnimation.save("movie.mp4", writer=writer)
```
